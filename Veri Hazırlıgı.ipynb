```python
import os
import shutil
import hashlib
from PIL import Image
```

```python
# Silinecek ikon boyutunu belirleyin (byte cinsinden)
ikon_boyut_limiti = 1 * 1024  # 10 KB

# Hedef dizin
dizin = "saglıklı_agac"  # Kendi dizin yolunuza göre değiştirin

# Dizin içindeki tüm .jpg dosyalarını kontrol edin
for dosya_adi in os.listdir(dizin):
    dosya_yolu = os.path.join(dizin, dosya_adi)

    # Sadece .jpg dosyalarını kontrol et
    if dosya_adi.lower().endswith('.jpg'):
        # Dosya boyutunu alın
        dosya_boyutu = os.path.getsize(dosya_yolu)

        # Eğer dosya boyutu limitin altındaysa, dosyayı sil
        if dosya_boyutu < ikon_boyut_limiti:
            os.remove(dosya_yolu)
            print(f"Silindi: {dosya_yolu}")

print("İşlem tamamlandı.")
```

```python
# Hedef dizin
dizin = "output_main"  # Kendi dizin yolunuza göre değiştirin

# Yeni dosya adı formatı
yeni_dosya_adi_format = "agac_{}.jpg"

# Dizin içindeki tüm .jpg dosyalarını kontrol edin
jpg_dosyalar = [f for f in os.listdir(dizin) if f.lower().endswith('.jpg')]

# Her bir dosyayı yeniden adlandır
for i, dosya_adi in enumerate(jpg_dosyalar, start=1):
    eski_dosya_yolu = os.path.join(dizin, dosya_adi)
    yeni_dosya_adi = yeni_dosya_adi_format.format(i)
    yeni_dosya_yolu = os.path.join(dizin, yeni_dosya_adi)

    os.rename(eski_dosya_yolu, yeni_dosya_yolu)
    print(f"'{eski_dosya_yolu}' yeniden adlandırıldı: '{yeni_dosya_yolu}'")

print("İşlem tamamlandı.")
```

```python
def resize_images(input_folder, output_folder, size):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
            img_path = os.path.join(input_folder, filename)
            img = Image.open(img_path)

            # RGBA modundaki görüntüleri RGB'ye dönüştür
            if img.mode == 'RGBA':
                img = img.convert('RGB')

            img = img.resize(size, Image.ANTIALIAS)
            output_path = os.path.join(output_folder, filename)
            img.save(output_path)
            print(f"Resized and saved {filename}")

# Kullanım
input_folder = 'main'  # Orijinal resimlerin bulunduğu klasör
output_folder = 'output_main'  # Ölçeklendirilmiş resimlerin kaydedileceği klasör
size = (224, 224)  # Ölçeklendirmek istediğiniz boyutlar (genişlik, yükseklik)

resize_images(input_folder, output_folder, size)
```

```python
def calculate_hash(file_path):
    hash_alg = hashlib.md5()
    with open(file_path, 'rb') as file:
        while chunk := file.read(8192):
            hash_alg.update(chunk)
    return hash_alg.hexdigest()

def move_duplicate_images(src_folder, dst_folder):
    if not os.path.exists(dst_folder):
        os.makedirs(dst_folder)

    hash_dict = {}
    for filename in os.listdir(src_folder):
        if filename.lower().endswith(".jpg"):
            file_path = os.path.join(src_folder, filename)
            file_hash = calculate_hash(file_path)

            if file_hash in hash_dict:
                dst_file = os.path.join(dst_folder, filename)
                counter = 1
                while os.path.exists(dst_file):
                    base_name, extension = os.path.splitext(filename)
                    dst_file = os.path.join(dst_folder, f"{base_name} ({counter}){extension}")
                    counter += 1

                shutil.move(file_path, dst_file)
                print(f"Moved: {file_path} to {dst_file}")
            else:
                hash_dict[file_hash] = file_path

# Kullanım
src_folder = "1"
dst_folder = "1_duplicate"
move_duplicate_images(src_folder, dst_folder)
```
